```python
import pandas as pd

df = pd.read_csv("Equal.csv", encoding="latin1", low_memory=False)

df.head()
```

```python
df = df.rename(columns={
    "ProductName": "product_name",
    "ProductPrice": "product_price",
    "Rate": "rating",
    "Review": "review_title",
    "Summary": "review_text",
    "Sentiment": "sentiment"
})

df.head()
```

```python
df["full_review"] = (
    df["review_title"].fillna("").astype(str) + " " +
    df["review_text"].fillna("").astype(str)
)

df[["product_name", "product_price", "rating", "full_review", "sentiment"]].head()
```

```python
df["rating"].value_counts()
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df["clean_price"] = (
    df["product_price"]
    .astype(str)
    .str.replace(r"[^0-9.]", "", regex=True)
)

df["clean_price"] = pd.to_numeric(df["clean_price"], errors="coerce")
```

```python
df[["product_price", "clean_price", "rating"]].head()
```

```python
pd.crosstab(df["rating"], df["sentiment"])
```

```python
df_clean = df[[
    "product_name",
    "clean_price",
    "rating",
    "full_review",
    "sentiment"
]].copy()

df_clean = df_clean.dropna(subset=["full_review", "sentiment", "rating"])

df_clean.head()
```

```python
df_clean.to_csv("marketmind_clean_reviews.csv", index=False)
```

# Phase 1 Dataset Selection

The earlier dataset with only `review` and `label` columns was not suitable for MarketMind AI because the reviews looked synthetic and lacked product context.

The new dataset `Equal.csv` is more suitable because it contains product name, price, rating, review title, review summary, and sentiment label.

Important columns:
- `ProductName`: product being reviewed
- `ProductPrice`: product price
- `Rate`: customer rating
- `Review`: short review title
- `Summary`: detailed customer feedback
- `Sentiment`: sentiment label

For ML preprocessing, `Review` and `Summary` will be combined into a new column called `full_review`.
